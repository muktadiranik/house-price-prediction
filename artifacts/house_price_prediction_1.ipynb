```python
def predict_price(Garage, Porch, PoolQC, Electrical, CentralAir, Heating, ExterCond, RoofStyle, HouseStyle, BldgType, Neighborhood, Utilities, Alley, Street, MSZoning, LotArea, TotalBsmtSF, GrLivArea, FullBath, BedroomAbvGr, KitchenAbvGr):
    X = np.zeros(len(x.columns))
    Garage_index = np.where(x.columns == Garage)[0][0]
    if Garage_index > 0:
        X[Garage_index] = 1
    Porch_index = np.where(x.columns == Porch)[0][0]
    if Porch_index > 0:
        X[Porch_index] = 1
    PoolQC_index = np.where(x.columns == PoolQC)[0][0]
    if PoolQC_index >= 0:
        X[PoolQC_index] = 1
    Electrical_index = np.where(x.columns == Electrical)[0][0]
    if Electrical_index >= 0:
        X[Electrical_index] = 1
    CentralAir_index = np.where(x.columns == CentralAir)[0][0]
    if CentralAir_index >= 0:
        X[CentralAir_index] = 1
    Heating_index = np.where(x.columns == Heating)[0][0]
    if Heating_index >= 0:
        X[Heating_index] = 1
    ExterCond_index = np.where(x.columns == ExterCond)[0][0]
    if ExterCond_index >= 0:
        X[ExterCond_index] = 1
    RoofStyle_index = np.where(x.columns == RoofStyle)[0][0]
    if RoofStyle_index >= 0:
        X[RoofStyle_index] = 1
    HouseStyle_index = np.where(x.columns == HouseStyle)[0][0]
    if HouseStyle_index >= 0:
        X[HouseStyle_index] = 1
    BldgType_index = np.where(x.columns == BldgType)[0][0]
    if BldgType_index >= 0:
        X[BldgType_index] = 1
    Neighborhood_index = np.where(x.columns == Neighborhood)[0][0] 
    if Neighborhood_index >= 0:
        X[Neighborhood_index] = 1
    Utilities_index = np.where(x.columns == Utilities)[0][0]
    if Utilities_index >= 0:
        X[Utilities_index] = 1
    Alley_index = np.where(x.columns == Alley)[0][0]
    if Alley_index >= 0:
        X[Alley_index] = 1
    Street_index = np.where(x.columns == Street)[0][0]
    if Street_index >= 0:
        X[Street_index] = 1
    MSZoning_index = np.where(x.columns == MSZoning)[0][0]
    if MSZoning_index >= 0:
        X[MSZoning_index] = 1
    X[83] = LotArea
    X[84] = TotalBsmtSF 
    X[85] = GrLivArea
    X[86] = FullBath 
    X[87] = BedroomAbvGr 
    X[88] = KitchenAbvGr
    return rlr_model.predict([X])[0]
```
